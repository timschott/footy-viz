```python
import pandas as pd
import matplotlib.pyplot as plt
#pip install mplsoccer
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
#Read in the data
df = pd.read_csv('mount_passes.csv')
```

```python
newDF = df[["id","location", "pass.end_location"]]
newDF.head
```

```python
def helper(x):

  x = x.replace("c", "")
  x = x.replace("(", "")
  x = x.replace(")", "")
  y = x.split(',')
  return y
```

```python
#x y endX endY
newDF["x"]=0
newDF["y"]=0
newDF["endX"]=0
newDF["endY"]=0

```

```python
for index, row in newDF.iterrows():
    y = helper(row['location'])
    z = helper(row['pass.end_location'])

    a = int(float(y[0].replace(" ","")))
    b = int(float(y[1].replace(" ","")))


    c = int(float(z[0].replace(" ","")))
    d = int(float(z[1].replace(" ","")))

    #print(str(a) + " " + str(b))
    newDF.at[index,"x"] = a
    newDF.at[index,"y"] = b

    newDF.at[index,"endX"] = c
    newDF.at[index,"endY"] = d

newDF.head()



```

```python
# now that all the data has been managed.

newDF['x'] = newDF['x']
newDF['y'] = newDF['y']
newDF['endX'] = newDF['endX']
newDF['endY'] = newDF['endY']
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

#this is how we create the pitch
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='grass', line_color='#c7d5cc', figsize=(25, 8),
              constrained_layout=False, tight_layout=True)

#Draw the pitch on the ax figure as well as invert the axis for this specific pitch
pitch.draw(ax=ax)
plt.gca().invert_yaxis()

#use a for loop to plot each pass
for x in range(len(newDF['x'])):
    if newDF['endX'][x]>80 and newDF['x'][x]<=119:
      #plt.plot((newDF['x'][x],newDF['endX'][x]),(newDF['y'][x],newDF['endY'][x]),color='#FFFFCC') 
      plt.arrow(newDF['x'][x], newDF['y'][x], newDF['endX'][x] - newDF['x'][x], newDF['endY'][x]- newDF['y'][x], head_width = 0.8, width=.05, ec = '#FFFFCC')
      plt.scatter(newDF['x'][x],newDF['y'][x],color='#FFE87C')

plt.title('Mason Mount\'s passes in the attacking third in Euro 2018 and World Cup 2020',color='white',size=20)
```
